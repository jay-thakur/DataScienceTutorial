# Why Numpy

You might be wondering, when we already have Python lists then why do we need  NumPy arrays ? 

To understand this, Let's see How Python stores an object in memory.

A Python object is actually a pointer to a memory location that stores all the details about the object, like bytes and the value. Although this extra information is what makes Python a dynamically typed language, it also comes at a cost which becomes apparent when storing a large collection of objects, like in an array.
Python lists are essentially an array of pointers, each pointing to a location that contains the information related to the element. This adds a lot of overhead in terms of memory and computation. And most of this information is rendered redundant when all the objects stored in the list are of the same type!

image.png

To overcome this problem, we use NumPy arrays -

* NumPy arrays are stored at one continuous place in memory unlike lists, so processes can access and manipulate them very efficiently. This behavior is called locality of reference in computer science. This is the main reason why NumPy is faster than lists. Also it is optimized to work with the latest CPU architectures.

image.png

* NumPy arrays contain only homogeneous elements, i.e. elements having the same data type. This makes it more efficient at storing and manipulating the array. This difference becomes apparent when the array has a large number of elements, say thousands or millions. 

* Also, with NumPy arrays, we can perform element-wise operations, something which is not possible using Python lists!

This is the reason why NumPy arrays are preferred over Python lists when performing mathematical operations on a large amount of data.

```python
import numpy as np
 
import time
import sys

arr = range(1000)
print("Size of each element of list in bytes: ",sys.getsizeof(arr))
print("Size of the whole list in bytes: ", sys.getsizeof(arr)*len(arr))
 
brr = np.arange(1000)
print("Size of each element of the Numpy array in bytes: ",brr.itemsize)
print("Size of the whole Numpy array in bytes: ", brr.size * brr.itemsize)
```

In this example, we can see that python list for 1000 elements takes 48000 butes whereas numpy array takes 8000 bytes.

```python
# create a python list.
my_list = list(range(1000000))

# create a Numpy array.
arr = np.arange(1000000)

%time for _ in range(10): my_list2 = [x ** 2 for x in my_list]

%time for _ in range(10): arr2 = arr ** 2
```

Here, we can see that Python list multiplication took `Wall time:3.04 s` whereas numpy array multiplication took `Wall time:22.3 ms`

# What is Difference between numpy array & a python list

1. The key difference between an array and a list is, **arrays are designed to handle vectorized operations while a python list is not**.

That means, if we apply a function it is performed on every item in the array, rather than on the whole array object.

Let’s suppose we want to add the number 2 to every item in the list.


```python
lst = [2, 3, 4, 5, 6]
# lst + 2  # error
```

That was not possible with a list. But we can do that on a `ndarray`.

```python
# Add 2 to each element of arr
arr = np.array([2, 3, 4, 5, 6])
arr + 2
```

2. Another characteristic is that, once a numpy array is created, we cannot increase its size. To do so, we will have to create a new array. But such a behavior of extending the size is natural in a list.

3. A numpy array must have all items to be of the same data type, unlike lists. This is another significant difference.

However, if we are uncertain about what datatype our array will hold or if we want to hold characters and numbers in the same array, we can set the dtype as **'object'**.

```python
#Creating an object array (to hold numbers as well as strings)

arr = np.array([1, 'Jay'], dtype='object')
print(arr)
print("Type of arr is : ", type(arr))

# Converting an array back to a list
arr = arr.tolist()
print(arr)
print("Type of arr is : ", type(arr))
```

To summarise, the main differences with python lists are:
* Arrays support vectorised operations, while lists don’t.
* Once an array is created, we cannot change its size. we will have to create a new array or overwrite the existing one.
* Every array has one and only one dtype. All items in it should be of that dtype.
* An equivalent numpy array occupies much less space than a python list of lists.

# Dimensions in Arrays ?

A dimension in arrays is one level of array depth.

## 0-D Arrays
0-D arrays, or Scalars, are the elements in an array. Each value in an array is a 0-D array.

```python
import numpy as np
arr = np.array(42)
print("arr: ", arr)
print("The type of arr: ", type(arr))
print("The dimension of arr:", np.ndim(arr))
```

## 1-D Arrays
An array that has 0-D arrays as its elements is called a uni-dimensional or 1-D array.
These are the most common and basic arrays.

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print(arr)
print("The dimension of arr:", np.ndim(arr))
```

## 2-D Arrays
An array that has 1-D arrays as its elements is called a 2-D array.
These are often used to represent matrix or 2nd order tensors. Numpy has a whole sub module dedicated towards matrix operations called numy.mat.

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr)
print("The dimension of arr:", np.ndim(arr))
```

## 3-D Arrays
An array that has 2-D arrays (matrices) as its elements is called 3-D array. These are often used to represent a 3rd order tensor.


```python
import numpy as np

arr = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(arr)
print("The dimension of arr:", np.ndim(arr))
```

# Broadcasting

Broadcasting is Numpy's terminology for performing mathematical operations between arrays with different shapes.


```python
arr1 = np.array([ [11, 12, 13], [21, 22, 23], [31, 32, 33] ])
arr2 = np.array([1, 2, 3])
print("Addition with broadcasting: ")
print(arr1 + arr2)
```

```python
print("Multiplication with broadcasting: ")
print(arr1 * arr2)
```

Broadcasting Rules

1. If all input arrays do not have the same number of dimensions, a “1” will be repeatedly prepended to the shapes of the smaller arrays until all the arrays have the same number of dimensions.

2. If an array has a size of 1 along a particular dimension it acts as if it has the size of the largest shape along that dimension by copying along that dimension.

```python
arr1 = np.array([[[1,3,4],[9,8, 5]],[[2,1,5],[8,6,2]]])
arr2 = np.array([6,5,3])
print(np.shape(arr1))
print(np.shape(arr2))
print(arr1 + arr2) # Elementwise addition with broadcasting
print("------------------\n", arr[0])
```

**Rule 2:** Array b is extended along the dimensions which have size 1 by copying the array as many time as needed until it is the same size as array a. This makes the size of array b to be (2,2,3).

After applying the broadcasting rules, the two arrays are the same size and elementwise add operation is performed.

# Expanding and Squeezing a NumPy Array

### Expanding a NumPy array

We can add a new axis to an array using the `expand_dims()` method by providing the array and the axis along which to expand.

```python
# expand dimensions
arr = np.array([1,2,3])
brr = np.expand_dims(arr, axis=0)
crr = np.expand_dims(arr, axis=1)
print('Original:','\n','Shape', arr.shape,'\n',arr)
print("------------------\n", arr[0])
print('Expand along columns:','\n','Shape',brr.shape,'\n', brr)
print("------------------\n", arr[0])
print('Expand along rows:','\n','Shape', crr.shape,'\n',crr)
```

### Squeezing a NumPy array

On the other hand, if we instead want to reduce the axis of the array, use the `squeeze()` method. It removes the axis that has a single entry. This means if we have created a `2x2x1` matrix, `squeeze()` will remove the third dimension from the matrix.

```python
# squeeze
arr = np.array([[[1,2,3], [4,5,6]]])
brr = np.squeeze(arr, axis=0)
print('Original','\n','Shape', arr.shape,'\n',arr)
print("------------------\n", arr[0])
print('Squeeze array:','\n','Shape', brr.shape,'\n',brr)
```

However, if we already had a 2 x 2 matrix, using `squeeze()` in that case would give us an error:

```python
# squeeze
arr = np.array([[1,2,3], [4,5,6]])
brr = np.squeeze(arr, axis=0)

print('Original','\n','Shape', arr.shape,'\n',arr)
print("------------------\n", arr[0])
print('Squeezed array:','\n','Shape', brr.shape,'\n',brr)
```
